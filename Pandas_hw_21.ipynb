# Задание 3

```python
rating = pd.read_csv('ratings.csv')
```

```python
lt = rating.groupby('userId').agg(['min', 'max'])['timestamp']
```

```python
lt['lifetime'] = lt['max'] - lt['min']
lt['lifetime'] = pd.to_datetime(lt['lifetime'], unit='s')
lt['lifetime2'] = lt.lifetime.dt.minute
ltt = lt.filter(items = ['userId', 'lifetime2'])
```

```python
cnt = rating.groupby('userId').agg(['count'])['timestamp']
cnt2 = cnt[(cnt['count'] > 100)]
```

```python
joined = cnt2.join(ltt, how='left')
#joined
```

```python
print ('среднее время жизни пользователей',round(sum(joined['lifetime2'])/len(joined['lifetime2'])),'минут')
```
